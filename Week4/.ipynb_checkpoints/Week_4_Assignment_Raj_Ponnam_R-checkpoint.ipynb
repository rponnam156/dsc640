```r
# Imports
library('magrittr')
library("dplyr")
```

```r
crime_rates_df <- read.csv2(paste("/Users/rajponnam/Documents/ds_course/dsc640/Week4/crimerates-by-state-2005.csv",sep=''), header=TRUE, sep=',', fill=TRUE, stringsAsFactors=FALSE)
head(crime_rates_df)
```

# Scatter plot

```r
options(repr.plot.width = 10, repr.plot.height = 10)

ggplot2::ggplot(data = crime_rates_df) +
  ggplot2::aes(x = murder, y = robbery) +
  ggplot2::geom_point(color = "#FF0000") +
  ggplot2::labs(title = "Correlation between Robbery and Murder",
       x = "Murder",
       y = "Robbery") +
  ggplot2::theme_grey()
```

# Bubble plot

```r
options(repr.plot.width = 10, repr.plot.height = 8)

ggplot2::ggplot(data = crime_rates_df) +
  ggplot2::aes(x = murder, y = robbery, color = total_crime, size = population) +
  ggplot2::geom_point() +
  ggplot2::labs(title = "Relationship between Murder, Robbery, Total Crime and Population") +
  ggplot2::theme_classic()
```

# Density plot

```r
# Density plot in ggplot2
ggplot(crime_rates_df, aes(x = x)) +
  geom_density(color = 4,
               fill = 4,
               alpha = 0.25) 
 
```
