```r
library('magrittr')

expenditures =  read.table(paste('/Users/rajponnam/Documents/ds_course/dsc640/Week3/expenditures.txt',sep='/'), header = TRUE, sep = '\t', dec = '.', fill = TRUE)
```

```r
head(expenditures)
```

# Treemap

```r
treemap::treemap(expenditures,index = c('category'),
                 vSize = 'expenditure',
                 title = 'Expenditure by Category')
```

# Area Chart

```r
expenditure_avg = expenditures %>%
dplyr::group_by(category) %>%
dplyr::summarize('Avg expenditure by category' = mean(expenditure))
head(expenditure_avg)
```

```r
ggplot2::ggplot(expenditure_avg, ggplot2::aes(x=category , y=`Avg expenditure by category`, group = 1)) +
ggplot2::geom_area( alpha=.5) +
ggplot2::geom_line() +
ggplot2::ggtitle('Average expenditure by category')
```
