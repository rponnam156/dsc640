```python
# read .txt file to pandas df
expenditures_df = pd.read_csv('/Users/rajponnam/Documents/ds_course/dsc640/Week3/expenditures.txt', sep = '\t', header=0)
```

```python
expenditures = expenditures_df.groupby(['category'])['expenditure'].sum().reset_index()

expenditures_by_year = expenditures_df.groupby(['year'])['expenditure'].sum().reset_index()

expenditures_df.head()
```

# Tree map

```python
squarify.plot(sizes=expenditures_by_year['expenditure'], label=expenditures_by_year['year'], alpha=.7 ) 
plt.axis('off') 
plt.show() 
```

# Area Chart

```python
# creating chart category
x = expenditures_df['year']
y = expenditures_df['category']
```

```python
# plot chart
plt.fill_between( x, y, color='orange', alpha=0.2)
plt.title('Total Expenditure by Year', loc='left')
plt.xlabel('Year')
plt.ylabel('Total Expenditure')
plt.plot(x, y, color='darkblue', alpha=0.6)
```

# Stacked Area Chart

```python
expenditures_plot = expenditures_df.loc[:, expenditures_df.columns != 'sex'].pivot(index='year', columns='category', values='expenditure')
expenditures_plot.reset_index(level=0, inplace=True)

expenditures_plot
```

```python

fig, ax = plt.subplots(1,1,figsize=(17, 10), dpi= 90)
columns = expenditures_plot.columns[2:]
labs = expenditures_plot.values.tolist()


labs = columns.values.tolist()
ax = plt.gca()
ax.stackplot(x, y, labels=labs, alpha=0.6)

ax.set_title('Total expenditure by year for each category', fontsize=20)
plt.xlabel('Year')
plt.ylabel('Total expenditure')

ax.legend(fontsize=12, ncol=1, loc = 'upper left')
plt.xticks(x[::6], fontsize=12, horizontalalignment='center')

plt.gca().spines["top"].set_alpha(0)
plt.gca().spines["bottom"].set_alpha(.2)
plt.gca().spines["right"].set_alpha(0)
plt.gca().spines["left"].set_alpha(.2)

plt.show()
```
