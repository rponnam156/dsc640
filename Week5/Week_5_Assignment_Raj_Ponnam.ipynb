# Countour plot

```python
%matplotlib inline

def sqrt_(x, y):
    return np.sqrt(x**2 + y**2)

x = np.array(ppg_df['FTM'])
y = np.array(ppg_df['FTA'])

X, Y = np.meshgrid(x, y)
Z = sqrt_(X, Y)

plt.figure()
cp = plt.contour(X, Y, Z)
plt.clabel(cp, inline=True, 
          fontsize=10)
plt.title('Contour Plot')
plt.xlabel('FTM')
plt.ylabel('FTA')
plt.show()
```
