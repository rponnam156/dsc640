```r
library('magrittr') 
library('ggmap')
library('ggplot2')
library('usmap')

costco_df <- read.csv('/Users/rajponnam/Documents/ds_course/dsc640/Week5/ex5-2/costcos-geocoded.csv')

costco_df_sum <- costco_df %>% 
  dplyr::mutate(state = State) %>% 
  dplyr::group_by(state) %>%
  dplyr::summarise(store_count = dplyr::n())
```

```r
ppg_df <- read.csv('/Users/rajponnam/Documents/ds_course/dsc640/Week5/ex5-2/ppg2008.csv')
```

# Heat Map

```r
heatmap(as.matrix(ppg_df[,2:ncol(ppg_df)]))
```

# Spatial chart

```r
options(repr.plot.width = 10, repr.plot.height = 5)

plot_usmap(data = costco_df_sum, values = "store_count", lines = "white") + 
  scale_fill_continuous(name = "Number of Costco Stores", label = scales::comma) + 
  theme(legend.position = "right")
```

# Contour plot

```r
ggplot(ppg_df, aes(x=FTM, y=FTA))+
  theme_bw()+
  geom_point(alpha=0.1, col='red')+
  geom_density2d(color='black')+
  ggtitle('Contour Plot')+
  theme(plot.title = element_text(hjust = 0.5))+
  labs(x='FTM', y='FTA')
```
